```python
import geopandas as gpd
import matplotlib.pyplot as plt
data = gpd.read_file("E:\Thesis\Data\cropped\districtratio.gpkg")
data.head()
```

```python
Total_pixel = data['totalpixel'].sum()
Total_light = data['ntl'].sum()
Total_road = data['road'].sum()
```

```python
mean_light= Total_light/Total_pixel
mean_road= Total_road/Total_pixel
```

```python
data['dev_road'] = data.road_ratio-mean_road
data['dev_light'] = data.ntl_ratio-mean_light
```

```python
data1=data.sort_values('dev_light')
data1
```

```python
fig, ax = plt.subplots(figsize=(20,20))
ax.plot(data1.ADM2_EN, data1.dev_light,'r',label='Light Deviation')
ax.plot(data1.ADM2_EN, data1.dev_road,'g',label='Road Deviation')
plt.rcParams.update({'font.size': 20})
plt.xticks(rotation= 90)
L=plt.legend()
plt.title("Light-Road deviation")
plt.savefig('E:/Thesis/Data/image/LightRoad.png')
```
