```python
from skimage import io
import scipy.stats
import os
def TestAndAppend(Image_x,Image_y):
    print(scipy.stats.pearsonr(Image_x.ravel(), Image_y.ravel()))
    print(scipy.stats.kendalltau(Image_x, Image_y))
    print(scipy.stats.spearmanr(Image_x, Image_y, axis=None))
def Plot(Image_x):
    plt.imshow(Image_x,cmap='pink')
    plt.show()
def plot(Image_x):
    plt.imshow(Image_x,cmap='terrain')
    plt.show()
```

```python
def TEST(i,j):
    Image_x  = io.imread(input_path+"tile_"+str(i)+"_"+str(j)+".tif",as_gray=True)
    Image_y = io.imread(input_path+"gtile_"+str(i)+"_"+str(j)+".tif",as_gray=True)
    TestAndAppend(Image_x,Image_y)
    Plot(Image_x)
    plot(Image_y)
```

```python
TEST(350,400)
TEST(250,800)
```
